```python
import keras_cv
import matplotlib.pyplot as plt

model = keras_cv.models.StableDiffusion(img_width=512, img_height=512)
```

```python
images = model.text_to_image("elvis riding a motorcycle", batch_size=1)
```

```python
plt.imshow(images[0])
```
